```python
#use glob to combine the Bike_Data into one csv file and load to data frame
glob_path = "Bike_Data"
all_files = glob.glob(glob_path+"/*.csv")
glob_df = pd.concat((pd.read_csv(f) for f in all_files))
```

```python
#count total rows and columns
glob_df.count()
```

```python
#rename columns
glob_df.columns = ["24-Hour_Passes", "24-Extra", "3-Day_Passes", "Annual", "7-Day","Col_1", "Total_1","Total_Trips_Since_Launch", "Date", "Miles_to_Date_One","Miles_to_Date", "Miles_Today", "Total_Members_1","Total_Members_All_Time", "Total_Sold","Trips_in_Past_24hrs"]
```

```python
#move Date column to front
date = glob_df['Date']
glob_df.drop(labels=['Date'], axis=1,inplace = True)
glob_df.insert(0, 'Date', date)
glob_df = glob_df.reset_index(drop=True)
glob_df.head()
```

```python
#Drop "Column 1" and "7-Day Passes columns"
glob_df.drop(['Annual','7-Day', 'Col_1', 'Total_1', 'Miles_to_Date_One', 'Total_Sold'], axis=1, inplace=True)
glob_df.head()
```

```python
#create function 'adder' that adds missing values to Total Trips and Miles to Date columns
def adder(y,z):
    
    
    #clean up columns
    holder = []
    for x in glob_df.iloc[:,y]:
        if isinstance(x, str) == True:
            x = x.replace(',','')
            x = x.strip(" \t")
            x=int(x)
            holder.append(x)
        else:
            holder.append(x)
            
    glob_df.iloc[:, y] = holder
    
    row_num = glob_df["Date"].count() 
    start = glob_df.iloc[0,y]
    
    #add 'trips in past 24 hours' column to create running total
    holder2 = []
    for x in range (1, row_num):
        start = start + glob_df.iloc[x,z]
        holder2.append(start)
    
    #put new values in dataframe
    glob_df.iloc[1:,y] = holder2
    
    
```

```python
#get missing values for column[3] form column[9]
adder(4,9)
```

```python
#get missing values for column[5]
adder(5,6)
```

```python
#check work
glob_df.tail()
```

```python
#combine columns
glob_df["24-Hour_Passes"] = glob_df["24-Hour_Passes"].combine_first(glob_df["24-Extra"])
glob_df["Total_Members_All_Time"] = glob_df["Total_Members_All_Time"].combine_first(glob_df["Total_Members_1"])
```

```python
glob_df.drop(['24-Extra', 'Total_Members_1'], axis=1, inplace=True)
```

```python
#save as csv
glob_df.to_csv("combined_daily.csv")
```

```python
#load NYC accident data
crash_df = pd.read_csv("Motor_Vehicle_Collisions_-_Crashes.csv")
```

```python
#check columns
pd.set_option('display.max_columns', 30)
crash_df.head()
```

```python
#select columns that only involve cyclists
crash_df =  crash_df[(crash_df['NUMBER OF CYCLIST INJURED'] > 0) | (crash_df['NUMBER OF CYCLIST KILLED'] > 0)]
```

```python
#sort by date
crash_df.sort_values('CRASH DATE', inplace=True)
crash_df = crash_df.reset_index(drop=True)
```

```python
#convert to datetime
crash_df['CRASH DATE'] = crash_df['CRASH DATE'].astype('datetime64[D]')
```

```python
# crash_df = crash_df.truncate('2015-01-01')
# crash_df = crash_df.reset_index(drop=True)
```

```python
#drop rows without latitude/longitude
crash_df = crash_df.dropna(subset=["LATITUDE"])
```

```python
crash_df.count()
```

```python
length2 = len(crash_df)
holder4 = []

#run nested for loop to create zip code column
for x in range(length2):
    result = search.by_coordinates(crash_df.iloc[x,4], crash_df.iloc[x,5], radius=5, returns=1)
    if result:
        for zipcode in result:
            ziplist = (zipcode.values())
            code = ziplist[0]
            holder4.append(code)
    else:
        holder4.append("none")
```

```python
len(holder4)
```

```python
crash_df["ZipCode"] = holder4
```
